```python
df=pd.read_csv('../DataSets/Abalone.csv',header=None)
df
```

```python
columns=['Sex','Length','Diameter','Height','Whole_Height',
        'Shucked_Weight','Viscera_Weight','Shell_Weight','Rings']
```

```python
dt = pd.get_dummies(data=df,columns=['Sex'],drop_first=True)
dt
```

```python
x=dt.drop(['Age'],axis=1)
```

```python
y=dt['Age']
```

```python
from sklearn.model_selection import train_test_split
xtr,xtes,ytr,ytes=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVR
model=SVR(C=30)
model.fit(xtr,ytr)
```

```python
ypr=model.predict(xtes)
ypr
```

```python
from sklearn.metrics import r2_score
svr=r2_score(ypr,ytes)
svr
```

```python
acc_score=[]
c_value=[0.00001,0.001,0.44,0.88,1,20,30]
for i in c_value:
    model_new=SVR(C=i)
    model_new.fit(xtr,ytr)
    ypred=model_new.predict(xtes)
    sc=r2_score(ypred,ytes)
    acc_score.append(sc)
```

```python
acc_score
```

```python
kernel=['linear','rbf','poly','sigmoid']
acc_sc=[]
for i in kernel:
    model_new2=SVR(kernel=i)
    model_new2.fit(xtr,ytr)
    y_pred=model_new2.predict(xtes)
    score=r2_score(ypred,ytes)
    acc_sc.append(score)
```

```python
report=pd.DataFrame({'Kernel':kernel,'Accuracy':acc_sc})
report
```
